```python
# =====================================
#       Correcting Column Names
# =====================================

print(f'Before: {df.columns.to_list()}\n')

df.columns = df.columns.str.strip().str.lower()

df = df.rename(columns={'campaignid':'campaign_id',
                        'campeigntag':'campaign_tag'})

print(f'After: {df.columns.to_list()}')
```

```python
# =====================================
#       Cleaning 'Spend' Column
# =====================================

dollar_mask = df['spend'].str.contains('\$')
print(f"Before:\n{df.loc[dollar_mask, ['spend']].head(3)}")

df['spend'] = df['spend'].str.replace(r'[-$, ]', '', regex=True)

print(f"After:\n{df.loc[dollar_mask, ['spend']].head(3)}")
```

```python
# =====================================
#     Correcting Categorical Typos 
# =====================================

print('Before:\n')
print(f"channel: {df['channel'].unique()}")
print(f"campaign_tag: {df['campaign_tag'].unique()}")

df['channel'] = df['channel'].replace({ 'Tik_Tok'   : 'TikTok',
                                        'Facebok'   : 'Facebook',
                                        'E-mail'    : 'Email',
                                        'Insta_gram': 'Instagram',
                                        'Gogle'     : 'Google Ads',
                                        np.nan      : 'Unknown' })

# Creating a map for 'campaign_tag' column according to 'channel' column.
tag_map = { 'TikTok'    : 'TI',
            'Facebook'  : 'FA',
            'Email'     : 'EM',
            'Instagram' : 'IN',
            'Google Ads': 'GO',
            'Unknown'   : 'Unknown' }

# First, replacing values and adding NaN values inplace of 'INVALID'.
df['campaign_tag'] = df['campaign_tag'].replace({ 'E-'    : 'EM',
                                                   'INVALID': np.nan, 
                                                   'XX'     : 'Unknown' })

# Second, using the map to fill tags where 'campaign_tag' was 'INVALID'.
df['campaign_tag'] = df['campaign_tag'].fillna(df['channel'].map(tag_map))

print('\nAfter:\n')
print(f"channel: {df['channel'].unique()}")
print(f"campaign_tag: {df['campaign_tag'].unique()}")
```

```python
# =====================================
#       Correcting Mixed Booleans
# =====================================

print('Before:\n')
print(f"active: {df['active'].unique()}")

df['active'] = df['active'].replace({ 'Y'    : True,
                                      '1'    : True,
                                      'Yes'  : True,
                                      'TRUE' : True,
                                      'No'   : False,
                                      '0'    : False,
                                      'FALSE': False}).astype('bool')

print('\nAter:\n')
print(f"active: {df['active'].unique()}")
```

```python
# =====================================
#          Changing Data Types
# =====================================
print('Before:\n')
df.info()

df = df.astype({ 'channel'     : 'category',
                 'campaign_tag': 'category' })  # To reduce memory usage.

def parse_date(x):                                                       # 1st method if you wanna parse each date format.
    for fmt in ("%m/%d/%Y %H:%M", "%m/%d/%Y", "%d/%m/%Y"):
        try:
            return pd.to_datetime(x, format=fmt)
        except:
            continue
    return pd.NaT

df['start_date'] = df['start_date'].apply(parse_date)

# df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')   # 2nd method if you want NaT.

df['end_date'] = pd.to_datetime(df['end_date'])

df['spend'] = pd.to_numeric(df['spend'], errors='coerce')

print('\nAfter:\n')
df.info()
```

```python
# ============================================
#  data integrity: 'Start_Date' vs 'End_Date'
# ============================================

# Creating a condtion mask.
mask = df['end_date'] < df['start_date']

# We have some records where end_date is smaller than start_date. Which is not right.
print('Before:\n')
print(df.loc[mask, ['start_date','end_date']].head(3))

# Assuming that end_date for these campaigns ends at a 30 days interval.
df.loc[mask, 'end_date'] = df.loc[mask, 'end_date'] + pd.Timedelta(days=30)

print('\nAfter:\n')
print(df.loc[mask, ['start_date','end_date']].head(3))
```

```python
# =====================================
#   Handling Outliers (Winsorization)
# =====================================

Q1 = df['spend'].quantile(q=0.25)
Q3 = df['spend'].quantile(q=0.75)

IQR = Q3 - Q1

upper_limit = Q3 + (3 * IQR)

outlier_mask = df['spend'] > upper_limit
print(df.loc[outlier_mask, ['campaign_id','spend']].head(3))

# Now we can do two things:
#  - Impute upper_limit in place of outliers.
#  - Make 'is_outlier' column to tag these records.

# In this project I'm going with 'Imputation'.
df.loc[df['spend'] > upper_limit, 'spend'] = upper_limit

print('\nAfter:\n')
print(df.loc[outlier_mask, ['campaign_id','spend']].head(3))

# But this is how 'is_outlier' column can be made.
# df['is_outlier'] = df['spend'].apply(lambda x: True if x > upper_limit else False)
```
